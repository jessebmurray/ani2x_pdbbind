```python
from ani2x import get_pdb_df, get_df_gen
import pandas as pd
```

```python
df_gen = get_df_gen()
```

```python
df_gen
```

```python
df_hetatm = get_pdb_df(atom_kind='hetatm', df_gen=df_gen, file_type='protein')
df_protein = get_pdb_df(atom_kind='protein', df_gen=df_gen, file_type='protein')
```

```python
df_hetatm
```

```python
df_protein.rename(columns={'element_symbol': 'element', 'residue_name': 'molecule_name',
                'x_coord': 'x', 'y_coord': 'y', 'z_coord': 'z'}, inplace=True)
df_hetatm.rename(columns={'element_symbol': 'element', 'residue_name': 'molecule_name',
                'x_coord': 'x', 'y_coord': 'y', 'z_coord': 'z'}, inplace=True)
df_protein = df_protein.astype({'x': 'float32', 'y': 'float32', 'z': 'float32'})
df_hetatm = df_hetatm.astype({'x': 'float32', 'y': 'float32', 'z': 'float32'})
```

```python
df_bind = pd.concat([df_protein, df_hetatm])
df_bind = df_bind[df_bind.PDB_ID != '2W73']
df_bind.sort_values(by=['PDB_ID', 'atom_kind', 'atom_number'], inplace=True)
```
